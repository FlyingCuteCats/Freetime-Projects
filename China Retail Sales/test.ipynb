```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cngen = pd.read_csv("CN General Catalogues.csv")
cncpi = pd.read_csv("CN CPI index.csv")
cnspe = pd.read_csv("CN Specific Catalogues.csv")

cngen['Date'] = pd.to_datetime(cngen['Date'])
cncpi['Date'] = pd.to_datetime(cncpi['Date'])
cnspe.rename(columns={'日期': 'Date'}, inplace=True)
cnspe['Date'] = pd.to_datetime(cnspe['Date'])
```

```python
cnspe = cnspe.join(cncpi.set_index('Date'), on='Date', how='inner')

cnspe.columns
```

```python
plt.plot(cnspe['Date'], cnspe['服装类商品零售类值当期值(亿元)'], label='Clothing')
plt.plot(cnspe['Date'], cnspe['日用品类商品零售类值当期值(亿元)'], label='Daily Necessities')
plt.legend()
```

```python
plt.plot(cncpi['Date'], cncpi['CPIIndex of Clothes'], label='Clothing CPI Index (Jan 2020 as 100)')
plt.axhline(y=100, linestyle='--', alpha=0.5)
plt.legend()
```

```python
plt.plot(cnspe['Date'], cnspe['服装类商品零售类值当期值(亿元)']/cnspe['CPIIndex of Clothes'], label='Adjusted Clothing Retail Sales')
plt.legend()
```

```python
plt.plot(cnspe['Date'], cnspe['家用电器和音像器材类商品零售类值当期值(亿元)'], label='Home Appliances')
plt.plot(cnspe['Date'], cnspe['家具类商品零售类值当期值(亿元)'], label='Furniture')
plt.plot(cnspe['Date'], cnspe['建筑及装潢材料类商品零售类值当期值(亿元)'], label='Building Materials')
plt.legend()
```

```python
plt.plot(cnspe['Date'], cnspe['家用电器和音像器材类商品零售类值当期值(亿元)']-cnspe['家具类商品零售类值当期值(亿元)']-cnspe['建筑及装潢材料类商品零售类值当期值(亿元)'])
plt.title('Home Appliances minus Furniture and Building Materials')
```
